## Excercise 1

```python
import json
import pprint
def str_to_dict(log, encoding='utf-8'):
    purchases = open(log,encoding=encoding)
    purchases_dict = {}
    for i, line in enumerate(purchases):
        if i > 0:
            purchases_dict[json.loads(purchases.readline())['user_id']] \
            = json.loads(purchases.readline())['category'] # ключи и значения purchases_dict также вытаскиваем из json-словаря
    return purchases_dict
        
pprint.pprint(str_to_dict('purchase_log.txt'))
```

## Excercise 2

```python
def get_user_purchase_category(output_file):
    purchases = str_to_dict('purchase_log.txt') # читаем полностью в ОЗУ
    with open('visit_log.csv','r') as r: # читаем построчно
        with open(output_file,'w') as w:
            for i, line in enumerate(r):
                if i > 0:
                    if line.strip().split(',')[0] in purchases.keys(): # проверяем наличие id из лога визитов в логе покупок
                        output_line = line.strip() + ',' + purchases[line.strip().split(',')[0]] + '\n'
                        w.write(output_line)
                    
get_user_purchase_category('test_1.csv')
```
